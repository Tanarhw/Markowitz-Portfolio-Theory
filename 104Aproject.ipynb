In the equations above, we can see that the secant method produces a much faster solution to our equation $\mu_{p} = w_{1}\mu_{1} + w_{2}\mu_{2}$ when $w_{1} = (1-\alpha)$ and $w_{2} = (\alpha)$. We have denoted the expected return of our portfolio to be $\mu_{p}$ and $w_{1}$ to be the fraction of an allotted wealth to the first asset and $w_{2}$ to be the allotted wealth to the second asset. $\mu_{1},\mu_{2}$ are the resepctive expected returns of each asset, and we ultimately seek to find a proper allocation to satisfy a desired expected return of 8%. Those proper allocation wieghts are: $w_{1} = (1-\alpha) = (1-(-4.628) = (5.628)$ so 562.8% of the portfolio should be weighted to asset 1, and $w_{2} = (\alpha) = (-4.628)$ so -462.8% of the portfolio should be weighted to asset 2. A negative weighting means an investor should short sell the equity, in essence immediately selling shares of the purchased equity, in this case our Costco equity, and waiting to re-purchase when the price drops. We have a total investment value between the two equities of $w = 18706.37$. In our case, we would immediately purchase then sell $86,573.08$ dollars of Costco equity and use the sold funds from the short sell and pump them into the Apple equity to purchase and hold $105,279.45$ dollars of Apple equity. This is an extreme case of short selling and then holding with our given weights, but is the result of the Markowitz theory.

A negative covariance describes that the performance between two stocks is not correlated, and an increase in price in one tends to mean a decrease in price in the other whereas a positive covarience means the assets move in the same pricing direction.
